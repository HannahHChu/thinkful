```python
import numpy as np
import matplotlib.pyplot as plt
```

# Binomial Distribution

```python
# numpy.random.binomial(n, p, size=None)
```

where:
    n = int or array_like of ints
    p = Parameter of the distribution, >= 0 and <=1
    size = output shape

```python
db_bi = np.random.binomial(50, 0.5, 1000)
```

### Histogram

```python
plt.hist(db_bi, bins=20, color = 'c')
plt.show()
```

### Descriptive Histogram

```python
plt.hist(db_bi, bins=20, color = 'c')

# Add a vertical line at the mean.
plt.axvline(db_bi.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(db_bi.mean() + db_bi.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(db_bi.mean()-db_bi.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

# Additional Problem

1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

2. Add them together to create a third variable.

```python
dbn1=np.random.normal(5, 0.5, 1000)
dbn2=np.random.normal(10, 1, 1000)
dbn3=dbn1+dbn2
```

3. Graph the third variable using a histogram.
4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.
5. Evaluate the descriptive statistics against the data.

```python
plt.hist(dbn3, bins=20, color='y')
plt.show()
```

```python
mean = np.mean(dbn3)
sd = np.std(dbn3)
```

```python
print("mean=",mean)
print("sd=",sd)
```

```python
plt.hist(dbn3, color='y')
plt.axvline(x=mean,color='b')
plt.axvline(x=mean+sd, color='r', linestyle='dashed')
plt.axvline(x=mean-sd, color='r',linestyle='dashed')
plt.show()
```
